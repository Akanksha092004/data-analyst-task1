Netflix Movies and TV Shows



```python
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/sample_data/akanksha/netflix_dataset (1).csv")
```

before cleaning


```python
print("Before Cleaning:")
print(df.shape)
print(df.isnull().sum())
```

1. Remove missing values

```python
df = df.dropna()
```

 2. Remove duplicates


```python
df = df.drop_duplicates()
```

3. Clean text columns



```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.lower().str.strip()
```

4. Rename columns


```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
print(df['date_added'].head())
print(df['date_added'].dtype)
```

5. Convert date column



```python
# df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.head(5)
```

```python
print(df['date_added'].isnull().sum())
```

```python
df['date_added'] = df['date_added'].fillna(pd.Timestamp('2000-01-01'))
```

```python
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
print(df.dtypes)
```

```python
df['release_year'] = df['release_year'].astype(int)
```

 After cleaning


```python
print("\nAfter Cleaning:")
print(df.shape)
print(df.isnull().sum())
```

```python
print(df)
```

```python
df.to_csv("cleaned_Netflix_dataset.csv", index=False)
```
